```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Load Data

```python
pd.set_option('display.max_columns', 500)
df = pd.read_csv('wbc.csv')
df
```

Memisahkan Variable Yang Tidak Digunakan

```python
df = df.drop(['Unnamed: 32', 'id'], axis=1)
df

# Penjelasan kode diatas
# *drop : Merupakan function untuk melakukan penghapusan terhadap kolom/baris yang tidak diperlukan pada dataset
# ['*'] : Merupakan parameter yang digunakan untuk menentukan kolom/baris yang akan dihapus
# axis=1 : Merupakan parameter yang digunakan untuk menentukan arah penghapusan, 1 = kolom, 0 = baris
```

Melakukan proses encoding pada kolom Diagnosis

```python
le = LabelEncoder()

df['diagnosis'] = le.fit_transform(df['diagnosis'])
df

# Penjelasan kode diatas
# *.fit_transform : Merupakan function yang digunakan untuk melakukan encoding data pada kolom diagnosis menjadi biner
```

Melakukan proses standarisasi pada semua kolom numerik

```python
# Mengambil semua kolom yang memiliki tipe data int64 dan float64
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
# Penjelasan kode diatas
# *.select_dtypes : Merupakan function yang digunakan untuk melakukan pemilihan kolom berdasarkan tipe data tertentu
# include=[*] : Merupakan parameter yang digunakan untuk menentukan tipe data yang akan diambil
# .columns : Merupakan method yang digunakan untuk mengambil nama kolom dari dataset

# Melakukan standarisasi data numerik
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
# Penjelasan kode diatas
# StandardScaler.fit_transform() : Merupakan function yang digunakan untuk melakukan standarisasi data numerik
# scaler.fit_transform(kolom) : [kolom] merupakan parameter yang digunakan untuk menentukan kolom yang akan di standarisasi
df
```

```python
# Melakukan split stratified data menjadi data train dan data test
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

# Penjelasan kode diatas
# train_test_split : Merupakan function yang digunakan untuk melakukan split data menjadi data train dan data test
# test_size=0.2 : Merupakan parameter yang digunakan untuk menentukan ukuran data test, 0.2 maksudnya 20% dari data menjadi
                # data testing dan 80% menjadi data training
# startify=[*] : Untuk memastikan bahwa data yang diambil dari data training dan data testing memiliki proporsi yang sama


# Menampilkan masing masing ukuran data
print("Jumlah data asli:", df.shape[0], "\n")
print(f'Jumlah label data training: \n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data testing: \n{df_test.diagnosis.value_counts()}')
```
